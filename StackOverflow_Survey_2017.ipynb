Programming Language Popularity

```
l = (list(df['HaveWorkedLanguage'].apply(lambda x: str(x).split('; '))))
from itertools import chain
y = list(chain(*l))
languages = list(set(y))

d = {}
for i in languages:
  d[i] = df['HaveWorkedLanguage'].apply(lambda x: i in str(x).split('; ')).value_counts()[1]

d.pop('nan')
languagelist = sorted(d.items(), key=lambda kv: kv[1], reverse=True)[:5]
D = OrderedDict(languagelist)

# Plot the chart
plt.bar(range(len(D)), list(D.values()), align='center')
plt.xticks(range(len(D)), list(D.keys()))

plt.show()
```
